```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib widget
```

```python
plt.figure()
plt.plot(np.arange(5))
```

```python
data = pd.read_csv('data_uk_2.csv', skiprows=4)
data.dropna(inplace=True)
```

```python
data.head()
```

```python
dat = data['Period\\Unit:'].to_numpy()
dat[dat == '-'] = np.nan
dat = np.float64(dat)
data['Period\\Unit:'] = dat
data.dropna(inplace=True)
data['date'] = data.index
data.plot(x='date', y=['Period\\Unit:'])
```

```python
data.head()
```

```python
data.plot()

data_price = data['Period\\Unit:'].to_numpy()
max_val = np.max(data_price)
max_idx = np.argmax(data_price)
min_val = np.min(data_price)
min_idx = np.argmin(data_price)

print(f"Max val of {max_val:.3f} at date {max_idx}")
print(f"Min val of {min_val:.3f} at date {min_idx}")

plt.axhline(max_val, linestyle='dotted')
plt.axhline(min_val, linestyle='dotted')
```

```python
data['diff'] = data['Period\\Unit:'].diff(-1)
data.plot(y=['diff'])
```
